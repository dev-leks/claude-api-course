```python
def add_message(messages, role, text):
  messages.append({
    'role': role,
    'content': text
  })

def chat(messages):
  message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages
  )
  return message.content[0].text
```

```python
messages = []

add_message(messages, 'user', 'What is quantum computing? Answer in one sentence')

response = chat(messages)
add_message(messages, 'assistant', response)

add_message(messages, 'user', 'Write another sentence.')

response = chat(messages)
add_message(messages, 'assistant', response)

messages
```
